### 3.Perform Below Visualizations: Univariate Analysis,Bivariate Analysis and Multivariate Analysis

#### Univariate Analysis

```python
categories = list(df['embarked'].unique())
value_counts = df['embarked'].value_counts()
plt.pie(value_counts,labels=categories[:-1])
```

```python
plt.hist(df['age'],color='violet')
```

#### Bivariate Analysis

```python
survival_counts = df.groupby('pclass')['survived'].value_counts().unstack()
colors = ['tab:blue', 'tab:red']
survival_counts.plot(kind='bar', stacked=True,color=colors)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Count')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Fare by Passenger Class')
plt.show()
```

#### Multivariate Analysis

```python
sns.catplot(x='pclass', hue='sex', col='survived', data=df, kind='count')
plt.suptitle('Survival by Passenger Class and Gender')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue='pclass', data=df)
plt.title('Age vs. Fare (colored by pclass)')
plt.show()
```

### 5. Handle the Missing values.

```python
df['deck'].fillna("None", inplace=True)
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df
```

### 6. Find the outliers and replace the outliers

```python
z_threshold = 3
z_scores = (df['age'] - df['age'].mean()) / df['age'].std()
outliers = (np.abs(z_scores) > z_threshold)
df.loc[outliers,'age'] = df['age'].mean()
```

```python
df
```
